```python
import pandas as pd
import os
import sys
import numpy as np
sys.path.append('..')

from urlclustering.noise_word import digit_ratio
from urlclustering.toolkit.noise_sample import write_sample, read_sample

samples = read_sample('../data/samples_meregs.csv')
```

```python
def isdigital(word):
    if digit_ratio(word) < 0.8 and len(word) < 40 and not word.startswith('c_') and '%' not in word:
        return True
    return False

samples[samples.word.apply(isdigital)]
```

```python
samples.loc[samples.word.apply(isdigital), 'label'] = False
```

```python
samples.loc[samples.label.apply(lambda label: np.isnan(label)), 'label'] = True
```
